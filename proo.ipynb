```python
from faker import Faker
import random

faker = Faker()

customer_ids = []
for i in range(70000):
    id_letters = ''.join(faker.random_letters(length=2))
    id_numbers = str(random.randint(1000000, 9999999))
    customer_ids.append("{}-{}".format(id_letters, id_numbers[:random.randint(6, 7)]))

# print the first 10 customer IDs to check if they match the format
print(customer_ids[:10])
```

```python
Customer_ID = pd.DataFrame({'Customer_ID':customer_ids})
Customer_ID[['Customer_ID']].nunique()
Customer_ID = Customer_ID[['Customer_ID']].drop_duplicates()
Customer_ID = Customer_ID[:51290]
len(Customer_ID)
Customer_ID
```
