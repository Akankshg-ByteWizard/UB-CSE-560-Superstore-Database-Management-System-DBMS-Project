```python
Product = data[['Product ID','Product Name','Category','Sub-Category','Unit Price','Sales']]
Customer = data[['Customer Name','Customer ID','Country','State','Region','City','Phone_number','Email_ID']]
Order = data[['Order ID', 'Order Date','Customer ID','Ship Mode','Product ID','Category','Order Priority']]
Billing = data[['Order ID','Customer ID','Product ID','Quantity','Unit Price','Sales','Discount','Shipping Cost']]
Shipping = data[['Order ID','Order Date', 'Order Priority','Ship Date', 'Ship Mode','Country','State','Region','City']]
```

```python
Product = Product.copy()
```

```python
Product['Product Name'].replace(',','-', inplace=True)
```

```python
Product = Product.drop_duplicates(subset=["Product ID", "Product Name"])
Product['row_number'] = Product.groupby('Sub-Category').ngroup() + 298
```

```python
Product['Unit Price']= Product['Unit Price'].round(2)
```

```python
Product['Sales']= Product['Sales'].round(2)
```

```python
Product.to_csv('Product.csv', index=False)
```
